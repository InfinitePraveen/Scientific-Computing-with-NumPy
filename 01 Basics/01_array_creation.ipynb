```python
import numpy as np
```

### Creating Array from a list

```python
arr_1d = np.array([1, 2, 3, 4, 5])
print("1D Array:", arr_1d)

arr_2d = np.array([[1, 2, 3], [4, 5, 6]])
print("\n2D Array:", arr_2d)
```

### List vs NumPY Arrays

```python
py_list = [1, 2, 3]
print("Python list multiplication", py_list * 2)
```

```python
np_array = np.array([1, 2, 3]) # Element wise multiplication
print("Python array multiplication", np_array * 2)
```

```python
import time
start = time.time()
py_list = [i * 2 for i in range(100000)]
print("\nList operation time: ", time.time()-start)
```

```python
start = time.time()
np_array = np.arange(100000) * 2
print(np_array)
print("\nNumpy operation time: ", time.time()-start)
```

### Creating array from scratch

```python
zeros = np.zeros((3, 4))
print("Zeros array:\n", zeros)

ones = np.ones((5, 7))
print("\nOnes array:\n", ones)

full = np.full((3, 6), 7)
print("\nFull arrays:\n", full)

random = np.random.random((4, 5))
print("\nRandom arrays:\n", random)

sequence = np.arange(0, 10, 2)
print("\nSequence array:\n", sequence)
```

### Vector, Matrix and Tensor

```python
vector = np.array([1, 2, 3])
print("Vector: ", vector)

matrix = np.array([[1, 2, 3],
                   [4, 5, 6]])
print("\nMatrix: ", matrix)

tensor = np.array([[[1, 2], [3, 4]],
                   [[5, 6], [7, 8]]])
print("\nTensor:\n", tensor)
```

### Array Properties

```python
array_for_properties = np.array([[1, 2, 3],
                                 [4, 5, True]])
print("Shape: ", array_for_properties.shape)
print("Dimention: ", array_for_properties.ndim)
print("Size: ", array_for_properties.size)
print("Datatype: ", array_for_properties.dtype)
```

### Array reshaping

```python
array_for_reshaping = np.arange(12)
print("Original array: ", array_for_reshaping)

reshaped = array_for_reshaping.reshape(3, 4)
print("\nReshaped array: ", reshaped)

flattened = reshaped.flatten()
print("\nFlattened array: ", flattened)

# Ravel: Return view, instead of copy
raveled = reshaped.ravel()
print("\nRaveled array: ", raveled)

# Transpose
transpose = reshaped.T
print("\nTransposed array: ", transpose)
```
